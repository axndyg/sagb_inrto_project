```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Shots

```python
england = pd.read_csv('/Users/kar/Documents/sagb_inrto_project/srcd_data/events_England.csv')
england.head()
```

```python
england_shots = england[england["subEventName"] == "Shot"]
england_shots.to_csv("/Users/kar/Documents/sagb_inrto_project/England/Csvs/England_shots.csv", index=False)
england_shots.head()
```

```python
england_shots = england_shots.drop(columns=["tags", "positions", "subEventId", "id", "tagsList", "pos_orig_y", "pos_orig_x", "pos_dest_y", "pos_dest_x"])

# Check the updated dataframe
england_shots.head()
```

```python
england_shots = england_shots.drop(columns="eventName")
```

```python
england_shots.head()
```

```python
england_shots.loc[england_shots["matchPeriod"] == "2H", "eventSec"] += 45 * 60
england_shots.tail()
```

```python
england_shots = england_shots.drop(columns="matchPeriod")
```

```python
del england_shots[england_shots.columns[0]]
```

```python
england_shots.head()
```

```python
england_shots = pd.read_csv("/Users/kar/Documents/sagb_inrto_project/England/Csvs/England_shots.csv")
england_shots['eventMin'] = (england_shots['eventSec'] / 60).round(1)

england_shots.head()
```

```python
england_shots.to_csv("/Users/kar/Documents/sagb_inrto_project/England/Csvs/England_shots.csv", index=False)
```

```python
sns.set_theme(style="white")

plt.figure(figsize=(10,6))
sns.histplot(england_shots['eventMin'], bins=range(0, 100, 10), kde=False)

plt.xlabel('Minutes of the Game')
plt.ylabel('Number of Fouls')
plt.title('Distribution of Shots Over Time in England')

plt.show()
```
